```python
#II part. LDA topic modelling
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(third_list)

# Create LDA model
lda_model = LatentDirichletAllocation(n_components=10, random_state=42)
lda_model.fit(X)
# Get the top 10 words for each topic
feature = vectorizer.get_feature_names_out()
for i, topic in enumerate(lda_model.components_):
    print(f"Top 7 words for Topic #{i}:")
    print([feature[j] for j in topic.argsort()[-7:]])
    print("\n")
```

III. We can see that many words overlap between many topics like track(topic 1, 7), program (topic 5, 8). My clusters were wrong in the first assignment, so I believe this was a lot different from the previous analysis and more accurate.

IV. After preprocessing the reviews and then performing LDA topic modelling, the output topics are accurate compared to the huge size of data. We were able to rank keywords based on their probability of frequency. By this we can see some important features are are unique from each other.

V. Yes. Other than topic 3 and 8, all other topics relate to the core functionality of tracking workout routines. But Wger Workout Manager does not have a subscription model (as seem from the frequent words) like this competitor opp.
